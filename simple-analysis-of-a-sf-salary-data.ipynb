# Third: Let's ask what we want! And the code will make the data recognize the answer :)

```python
df['BasePay']
```

```python
#1-average Baesbay?
df['BasePay'].replace('Not Provided',0,inplace=True)
df['BasePay']=df['BasePay'].astype(float)
df['BasePay'].mean()
```

```python
df['OvertimePay']
```

```python
#2-what is the highest ammount of over overtimepay?
df['OvertimePay'].replace('Not Provided',0,inplace=True)#
df['OvertimePay']=df['OvertimePay'].astype(float)#
df['OvertimePay'].max()
```

```python
#3-what is the jobtitle of GARY JIMENEZ?
df[df['EmployeeName']=='JOSEPH DRISCOLL']['JobTitle']
```

```python
#4-How much does JOSEPH DRISCOLL make (including benefits)?
df[df['EmployeeName']=='JOSEPH DRISCOLL']['TotalPayBenefits']
```

```python
#5-What is the name of highest paid person (including benefits)?
df[df['TotalPayBenefits']==df['TotalPayBenefits'].max()]['EmployeeName']
```

```python
#6-Who are the employees who have debts to the company?
df[df['TotalPayBenefits']<0]
```

```python
#7-What was the average (mean) basePay of all employees per year? (2011-2014)?
df.groupby("Year").BasePay.mean()
#df.groupby(["Year"]).mean()["basePay"]
```

```python
#8-How many unique job titles are there?
df["JobTitle"].nunique()
```

```python
#9-what the unique job titles are there?
df["JobTitle"].unique()
```

```python
#10-What are the top 5 most common jobs?
df['JobTitle'].value_counts().head(5)
```

```python
#11-What are the less 5 most common jobs?
df['JobTitle'].value_counts().tail(5)
```

```python
#12-How many Job Titles were represented by only one person in 2013? 
sum(df[df["Year"]==2013]["JobTitle"].value_counts()==1)
#df[df['Year']==2013]["JobTitle"].nunique()
```

```python
#13-How many people have the word "Chief" in their job title?
def chiefString(x):
    if "chief" in x.lower():
        return True
    else:
        return False

sum(df["JobTitle"].apply(lambda x: chiefString(x)))
```

```python
#14-Is there a correlation between length of the Job Title string and Salary?
df['TitleLength'] = df['JobTitle'].apply(len)
df[['TitleLength','TotalPayBenefits']].corr()
```
